```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get(i, headers={'User-Agent': '/'})
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
c = soup.findAll(attrs = {"class": "article-post"})
c
```

```python
content = c[0].text.replace("\n"," ")
```

```python
content
```

```python
title = soup.find(attrs = {"class": "posttitle"}).text
title
```

```python
data = [[i, title, content]]
```

```python
data
```

```python
url = ['https://practicaldatascience.co.uk/data-science/how-to-use-the-dropbox-api-with-python', 'https://practicaldatascience.co.uk/data-science/how-to-read-an-xml-feed-into-a-pandas-dataframe',
      'https://practicaldatascience.co.uk/data-engineering/how-to-backup-a-mysql-database-with-mysqldump-ssh-and-scp', 'https://practicaldatascience.co.uk/data-science/how-to-use-the-pandas-apply-function',
      'https://practicaldatascience.co.uk/data-science/how-to-create-descriptive-statistics-using-the-pandas-describe-function', 'https://practicaldatascience.co.uk/machine-learning/16-how-to-create-google-search-console-time-series-forecasts-using-neural-prophet', 'https://practicaldatascience.co.uk/data-science/16-python-web-scraping-projects-for-ecommerce-and-seo']
for i in url:
    page = requests.get(i, headers={'User-Agent': '/'})
    if page.status_code == 200:
        print("Data Fetched Successfully:", i)
        soup = BeautifulSoup(page.content, 'html.parser')
        c = soup.findAll(attrs = {"class": "article-post"})
        content = c[0].text.replace("\n"," ")
        title = soup.find(attrs = {"class": "posttitle"}).text
        data.insert(1, [i, title, content])
    else:
        print('url not found', i)
df1 = pd.DataFrame(data, columns = ['URL', 'Title', 'Content'])
df1.to_csv('urlcontent.csv')
```
