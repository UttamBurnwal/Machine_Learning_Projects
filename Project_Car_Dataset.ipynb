**Project- Data Analysis (Car Dataset)**

1) Finding all Null Values in the dataset. If there are any null value in any column, then I will fill it with the mean of that column.

2) Checking the different types of Make in the dataset. And, what is the count (occurrence) of each Make in the data ?

3) Showing all the records where Origin is Asia or Europe.

4) Removing all the records (rows) where Weight is above 4000.

5) Increasing all the values of 'MPG_City' column by 3.

```python
import pandas as pd
```

```python
df = pd.read_csv('Cars Data.csv')
```

```python
# Finding all Null Values in the dataset. 

df.isnull().sum()
```

```python
df.dtypes
```

```python
# Filling the null values with the mean and mode of the column. For object type columns, I used mode and for float type datatype I used mean.

df['Make']= df['Make'].fillna(df['Make'].mode()[0])
df['Model']= df['Model'].fillna(df['Model'].mode()[0])
df['Type']= df['Type'].fillna(df['Type'].mode()[0])
df['Origin']= df['Origin'].fillna(df['Origin'].mode()[0])
df['DriveTrain']= df['DriveTrain'].fillna(df['DriveTrain'].mode()[0])
df['MSRP']= df['MSRP'].fillna(df['MSRP'].mode()[0])
df['Invoice']= df['Invoice'].fillna(df['Invoice'].mode()[0])
```

```python
df['EngineSize'] = df['EngineSize'].fillna(df['EngineSize'].mean())
df['Cylinders'] = df['Cylinders'].fillna(df['Cylinders'].mean())
df['Horsepower'] = df['Horsepower'].fillna(df['Horsepower'].mean())
df['MPG_City'] = df['MPG_City'].fillna(df['MPG_City'].mean())
df['MPG_Highway'] = df['MPG_Highway'].fillna(df['MPG_Highway'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
df['Wheelbase'] = df['Wheelbase'].fillna(df['Wheelbase'].mean())
df['Length'] = df['Length'].fillna(df['Length'].mean())
```

```python
#  Checking the different types of Make in the dataset

df['Make'].unique()
```

```python
#  count (occurrence) of each Make in the data 

df['Make'].value_counts()
```

```python
# Records where Origin is Asia or Europe

df[(df['Origin']=='Asia') | (df['Origin']== 'Europe')]
```

```python
df[df['Origin'].isin(['Asia', 'Europe'])]
```

```python
# Removing all the records (rows) where Weight is above 4000.

df[~(df['Weight']> 4000)]
```

```python
# Increasing all the values of 'MPG_City' column by 3

df['MPG_City']= df['MPG_City'].apply(lambda x:x+3)
```
