```python
# Weather is exactly clear

df['weather'].value_counts()
```

```python
df[df.weather=='Clear']
```

```python
df.groupby('weather').get_group('Clear')
```

```python
# Wind Speed exactly 4 km/hr

df[df['ws']==4]
```

```python
# Finding the null values

df.isnull().sum()
```

```python
# Renaming the Weather column
# I had already renamed all the columns but I will do it again
df.rename(columns={'weather': 'Weather_Condition'}, inplace =True)
```

```python
# Mean of visibility column

df.vis.mean()
```

```python
# Standard Deviation of Pressure

df.pressure.std()
```

```python
# Variance of Relative Humidity

df.rhum.var()
```

```python
# All instances where there is only snow 

df['Weather_Condition'].value_counts()
```

```python
df[df['Weather_Condition']=='Snow']
```

```python
# All instances where snow is recorded (with other weather conditions as well)
df[df['Weather_Condition'].str.contains('Snow')]
```

```python
# All instances when 'Wind Speed is above 24' and 'Visibility is 25'

df[(df['ws']>  24) & (df['vis']==25)]
```

```python
# Mean Value of each column against Weather Condition

df.groupby('Weather_Condition').mean()
```

```python
# Max and Min Value of each column against Weather Condition

df.groupby('Weather_Condition').max()
```

```python
df.groupby('Weather_Condition').min()
```

```python
# Records where  Weather Condition is Fog

df[df['Weather_Condition'] == 'Fog']
```

```python
# Instances  where Weather is clear and Visibility is above 40

df[(df['Weather_Condition']=='Clear') & (df['vis'] >40)]
```

```python
  # Instances when 'Weather is Clear' and 'Relative Humidity is greater than 50' or Visibility is above 40'

df[(df['Weather_Condition']  =='Clear') & (df['rhum']>50) | (df['vis']>40)]
```
